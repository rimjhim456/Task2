```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y, color='blue', linestyle='--', marker='o', label='y=2x')
plt.title("Line Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ['America','japan','singapore','china']
values = [956,856,562,965]

plt.bar(categories, values, color='green', width=0.5)
plt.title("VAT")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
x = [0.5,0.7,0.8,0.07,00.6,0.9,0.05,0.6]
y = [9.9,8.6,8.7,8.8,1.00,8.06,1.03,8.07]

plt.scatter(x, y, color='black', s=100, alpha=0.9)
plt.title("Scatter Plot ")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
sizes = [20, 30, 25, 25]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
colors = ['gold','lightblue','lightgreen','pink']
explode = [0,0.1,0,0]

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=120, explode=explode)
plt.title("Pie Chart")
plt.show()
```

```python
import seaborn as sns
import numpy as np

data = np.random.randn(1000)
sns.histplot(data, bins=30, kde=True, color='purple')
plt.title("Distribution Plot with Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data = [0.2,0.0025,0.2569,0.2316,0.00258,0.1569,0.25]
sns.boxplot(data=data)
plt.title("Box Plot Example")
plt.xlabel("Value")
plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.title("Violin Plot Example")
plt.show()
```
